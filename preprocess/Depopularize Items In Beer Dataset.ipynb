```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/beer/' + 'Data.csv')
```

```python
len(df[df['Binary'] == 1])
```

```python
item_count_df = df.groupby(['ItemIndex'])['Binary'].sum().to_frame().reset_index().sort_values(by='Binary', ascending=False)
```

```python
item_count_df
```

```python
item_count_df['Binary'].tolist()
```

```python
zero_items = item_count_df[item_count_df['Binary'] == 0]['ItemIndex'].values
```

```python
len(zero_items)
```

```python
zero_items
```

```python
more_than_200_items = item_count_df[item_count_df['Binary'] > 200]['ItemIndex'].values
```

```python
more_than_200_items
```

```python
len(more_than_200_items)
```

```python
delete_items = np.concatenate((zero_items, more_than_200_items), axis=None)
```

```python
len(delete_items)
```

```python
filtered_df = df[~df.ItemIndex.isin(delete_items)]
filtered_df
```

```python
print(filtered_df['keyphrases_indices_length'].sum())
print(len(filtered_df[filtered_df['keyphrases_indices_length'] > 0])/len(filtered_df))
print(filtered_df['keyphrases_indices_length'].sum() / len(filtered_df))
```

```python
ITEM_ID = 'beer/name'
USER_ID = 'user/profileName'
RATING = 'review/overall'
BINARY_RATING = 'Binary'
REVIEW_TEXT = 'review/text'
TIMESTAMP = 'review/timeUnix'
DATA_PATH = '../data/Beer/advocate/'
DATA_NAME = 'RefinedRawData.csv'
```

```python
filtered_df['UserIndex'] = filtered_df[USER_ID].astype('category').cat.rename_categories(range(0, filtered_df[USER_ID].nunique()))
filtered_df['ItemIndex'] = filtered_df[ITEM_ID].astype('category').cat.rename_categories(range(0, filtered_df[ITEM_ID].nunique()))
filtered_df['Binary'] = (filtered_df[RATING] == 5)*1

df_user_name = filtered_df[['UserIndex',USER_ID]]
df_item_name = filtered_df[['ItemIndex',ITEM_ID]]

total_review = len(filtered_df)

print("Number of User: {0}".format(df_user_name['UserIndex'].nunique()))
print("Number of Item: {0}".format(df_item_name['ItemIndex'].nunique()))
print("Number of Positive Review: {0}".format(sum(filtered_df['Binary'].tolist())))
# print("Number of key Phrase: {0}".format(len(df_keyphrases)))
print("Number of reviews: {0}".format(total_review))
```

```python
dataPath = '../data/beer/'

filtered_df.to_csv(dataPath+'Data.csv')
df_user_name.to_csv(dataPath+'UserIndex.csv')
df_item_name.to_csv(dataPath+'ItemIndex.csv')
```
