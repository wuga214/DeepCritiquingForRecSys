```python
import re
import itertools
from collections import Counter
import pandas as pd

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tag import PerceptronTagger
from nltk.data import find
from nltk.tokenize import RegexpTokenizer
from nltk import sent_tokenize

from keras.preprocessing.text import one_hot
from keras.preprocessing.text import text_to_word_sequence
from keras.preprocessing.text import Tokenizer

import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
itemID = 'asin'
userID = 'reviewerID'
rating = 'overall'
reviewText = 'reviewText'
columnToDrop = ['reviewTime','helpful','reviewerName', 'summary']
dataPath = 'data/baby/'
dataName = 'Baby_5.json'
```

```python
df = pd.read_json(dataPath+dataName, lines=True).drop(columnToDrop, axis=1)
df.head()
```

```python
subjective = ['i', 'my', 'me', 'mine', 'we', 'our', 'he', 'she', 'him', 'her']
def highQualitySentences(text):
    text = text.lower()
    sentences = sent_tokenize(text)
    for i in reversed(range(len(sentences))):
        if any(word in sentences[i].split(' ') for word in subjective):
            del sentences[i]
    return ' '.join(sentences[:3])
```

```python
df[reviewText][0]
```

```python
highQualitySentences(df[reviewText][0])
```

```python
df[reviewText] = df[reviewText].apply(lambda x: highQualitySentences(x))
```
